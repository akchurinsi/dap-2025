2. Создать 8x8 матрицу и заполнить её единицами и нулями в шахматном порядке

````python
import numpy as np
matrix = np.zeros((8, 8), dtype=int)
for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 0:
            matrix[i, j] = 1
print("8x8 матрица в шахматном порядке:")
print(matrix)
````

3. Дан массив из 11 элементов, поменять знак у элементов, значения которых между 3 и 8

````python
import numpy as np
array = np.random.randint(-10, 11, size=11)
print("Исходный массив:")
print(array)
for i in range(len(array)):
    if 3 <= array[i] <= 8:
        array[i] = -array[i]
print("Массив после изменения знака:")
print(array)
````

4. Найти среднее заначение строки матрицы и вычесть его из каждого елемента строки

````python
import numpy as np
z = np.random.rand(5, 10)
row_means = np.mean(z, axis=1, keepdims=True)
centered_z = z - row_means
print("Исходная матрица z:")
print(z)
print("\nСредние значения строк:")
print(row_means.flatten())
print("\nМатрица после вычитания средних из элементов строк:")
print(centered_z)
````

6. Дан массив 10x2 подеставлющий собой точки в декартовой системе координат, преобразовать в полярную

````python
import numpy as np
z = np.random.random((10, 2))
print("Исходный массив z (декартовы координаты x, y):")
print(z)
# Вычислить радиус r = sqrt(x^2 + y^2)
r = np.sqrt(z[:, 0]**2 + z[:, 1]**2)
# Вычислить угол theta = atan2(y, x) в радианах
theta = np.arctan2(z[:, 1], z[:, 0])
# Создать новый массив с полярными координатами (r, theta)
polar_z = np.column_stack((r, theta))
print("\nМассив в полярных координатах (r, theta в радианах):")
print(polar_z)
````

8. Дан массив, добавить 1 к каждому элементу с индексом, заданным в другом массиве. Добавление производиться только 1 раз

9. Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением

````python
import numpy as np

z = np.array([1, 2, 3, 4, 5])

print("Исходный вектор z:")
print(z)
k = 3
step = k + 1
new_length = len(z) + (len(z) - 1) * k
new_z = np.zeros(new_length)
new_z[::step] = z
print("\nНовый вектор с тремя нулями между каждым значением:")
print(new_z)
````

10. Найти ближайшее к заданному значению число в заданном массиве

````python
import numpy as np
z = np.arange(100)
v = np.random.uniform(0, 100)
print("Массив z:")
print(z)
print(f"\nЗаданное значение v: {v}")
differences = np.abs(z - v)
closest_index = np.argmin(differences)
closest_value = z[closest_index]
print(f"\nБлижайшее значение: {closest_value} (индекс {closest_index})")
````
